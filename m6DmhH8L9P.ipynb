**Problem #5** Using the given *Residue* list write a short code fragment to print all possible 10-peptide chains with a molecular weight of 1024 Daltons. More credit will be given for methods with effective pruning. (**Warning**: The list is long, and you probably do not have enough time to run it.)

````python
Residues = {'A':  71, 'C': 103, 'D': 115, 'E': 129,
            'F': 147, 'G':  57, 'H': 137, 'I': 113,
            'K': 128, 'L': 113, 'M': 131, 'N': 114,
            'P':  97, 'Q': 128, 'R': 156, 'S':  87,
            'T': 101, 'V':  99, 'W': 186, 'Y': 163
}

def genPep(maxWeight, prefix = ''):
    weight = (sum([Residues[peptide] for peptide in prefix]))
    if (prefix == ''):
        global peptideList
        peptideList = []
    if (weight == maxWeight and len(prefix) == 10):
        peptideList.append(''.join(sorted(prefix)))
        return 
    elif(weight > maxWeight):
        # PRUNE
        return
    elif(len(prefix) > 10):
        # PRUNE
        return
    else:
        for key in Residues.keys():
            if (weight + Residues[key] <= maxWeight):
                # PRUNE
                genPep(maxWeight, prefix+key)
                
genPep(1024)
# Not sure if you want a set or list since you want all possible
# print len(set(peptideList)), sorted(set(peptideList))
print len(peptideList), sorted(peptideList)
````
