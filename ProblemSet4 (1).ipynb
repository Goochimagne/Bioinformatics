**Problem #4: Programming Problem**

Write a function to compute the LCP array for the implicit suffix array of a given BWT string.

```python
def computeLCP(bwt):
    t = inverseBWT(bwt)
    sa = suffixArray(t)
    N = len(t)
    lcp = [0 for i in xrange(N)]
    for i in xrange(1,N):
        matches = 0
        x = sa[i]
        y = sa[i-1]
        while t[x] == t[y]:
            matches += 1
            x += 1
            y += 1
            if (x >= N) or (y >= N):
                break
        lcp[i] = matches
    return lcp
```
